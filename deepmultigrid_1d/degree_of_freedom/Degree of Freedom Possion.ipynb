```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

def interpolate(m):
    INT = np.zeros([2**(m+1)-1,2**m-1]);

    for i in range(2**m-1):
        INT[2*i,i]=1/2;
        INT[2*i+1,i]=1;
        INT[2*i+2,i]=1/2;

    return INT;

def restriction(m):
    RESTRICT=np.zeros([2**m-1,2**(m+1)-1]);

    for i in range(2**m-1):
        RESTRICT[i,2*i]=1/4;
        RESTRICT[i,2*i+1]=1/2;
        RESTRICT[i,2*i+2]=1/4;
    return RESTRICT;
    
def Possion(m):

    A = (2**(2*m+3))*np.eye(2**(m+1)-1);

    for i in range(2**(m+1)-2):
        A[i,i+1]=-(2**(2*m+2));
        A[i+1,i]=-(2**(2*m+2)); 

    return A;

def Helmholtz(m):

    f = lambda x: x**2;

    H = Possion(m)
    for i in range(len(H)):
        H[i, i] = H[i, i] - f(i/len(H))**2; 
    
    return H;

def Local_optimizer(m, A0, P0, R0, w0, NUM_EPOCHS):
    s1 = s2 = 1;
    A = tf.constant(A0, dtype = tf.float32);
    P = tf.Variable(initial_value = P0, dtype = tf.float32);
    R = tf.Variable(initial_value = R0, dtype = tf.float32);
    w = tf.Variable(initial_value=w0, dtype = tf.float32);
    M10 = tf.constant(np.diag(np.diag(A0)), dtype = tf.float32);
    M1 = w**(-1)*M10;
    K1 = M1 - A;
    M20 = tf.constant(np.diag(np.diag(A0)), dtype = tf.float32);
    M2 = w**(-1)*M20;
    K2 = M2 - A;
    M2K2s2 = tf.pow(tf.matmul(tf.matrix_inverse(M2), K2), s2);
    M1K1s1 = tf.pow(tf.matmul(tf.matrix_inverse(M1), K1), s1);
    RAP = tf.matmul(tf.matmul(R, A), P);
    IPRAPRA = tf.eye(2**(m+1)-1) + tf.matmul(tf.matmul(tf.matmul(P, tf.matrix_inverse(RAP)), R) ,A);
    C = M2K2s2*IPRAPRA*M1K1s1;
    Cnorm = tf.norm(C,ord=2);
    loss = Cnorm;
    optimizer = tf.train.GradientDescentOptimizer(0.01);
    train_op = optimizer.minimize(loss);
    sess = tf.Session();
    init = tf.global_variables_initializer();
    sess.run(init);
    Norm = [];
    for i in range(NUM_EPOCHS):
        sess.run(train_op);
        cost = sess.run(Cnorm);
        Norm.append(cost);

    R = sess.run(R);
    P = sess.run(P);
    w = sess.run(w);
    norm = Norm[-1];
    return R, P, w, norm

NUM_EPOCHS_SGD = 10;
step = 1/(2**3);
m=3;
t = 20;
S1 = 5;
S2 = 5;

A0 = Possion(m);
A1 = Helmholtz(m);
R0 = restriction(m);
P0 = interpolate(m);
w0 = 2/3;
accept_norm = 0.1;
M = (1 - step)*A0 + step*A1;
L = step;
norm = 1;
while L <  1:        
        Rnew, Pnew, wnew, norm = Local_optimizer(m, M, P0, R0, w0, NUM_EPOCHS_SGD);
        if norm < accept_norm:
            R0 = Rnew;
            P0 = Pnew;
            w0 = wnew;
            M = M - step*A0 + step*A1;
            L = L + step
            step = step * 2;
        else:
            M = M + step*A0 - step*A1
            step = step/2;
            M = M - step*A0 + step*A1;
            NUM_EPOCHS_SGD = NUM_EPOCHS_SGD*2;
Raw_Restriction = restriction(m)
Raw_Interpolate = interpolate(m)
Optimized_Restriction = R0
Optimized_Interpolate = P0


print('----------------------------------Raw_Restriction----------------------------------')
n = Raw_Restriction.shape[1];
for i in range(n):
    X = np.zeros([n,1]);
    X[i] = 1;
    OUTPUT = np.dot(Raw_Restriction,X);
    U = np.linspace(0,1,len(OUTPUT));
    print('X is');
    print(X);
    print('OUTPUT is');
    print(OUTPUT);
    plt.figure(i)
    plt.plot(U, OUTPUT,label='Raw_Restriction');
    plt.xlabel('u')
    plt.ylabel('OUTPUT')
    plt.legend(loc='upper left')
    plt.show()

print('----------------------------------Raw_Interpolate----------------------------------')
n = Raw_Interpolate.shape[1];
for i in range(n):
    X = np.zeros([n,1]);
    X[i] = 1;
    OUTPUT = np.dot(Raw_Interpolate,X);
    U = np.linspace(0,1,len(OUTPUT));
    print('X is');
    print(X);
    print('OUTPUT is');
    print(OUTPUT);
    plt.figure(1)
    plt.plot(U, OUTPUT,label='Raw_Interpolate');
    plt.xlabel('u')
    plt.ylabel('OUTPUT')
    plt.legend(loc='upper left')
    plt.show()

print('----------------------------------Optimized_Restriction----------------------------------')
n = Optimized_Restriction.shape[1];
for i in range(n):
    X = np.zeros([n,1]);
    X[i] = 1;
    OUTPUT = np.dot(Optimized_Restriction,X);
    U = np.linspace(0,1,len(OUTPUT));
    print('X is');
    print(X);
    print('OUTPUT is');
    print(OUTPUT);
    plt.figure(1)
    plt.plot(U, OUTPUT,label='Optimized_Restriction');
    plt.xlabel('u')
    plt.ylabel('OUTPUT')
    plt.legend(loc='upper left')
    plt.show()

print('----------------------------------Optimized_Interpolate----------------------------------')
n = Optimized_Interpolate.shape[1];
for i in range(n):
    X = np.zeros([n,1]);
    X[i] = 1;
    OUTPUT = np.dot(Optimized_Interpolate,X);
    U = np.linspace(0,1,len(OUTPUT));
    print('X is');
    print(X);
    print('OUTPUT is');
    print(OUTPUT);
    plt.figure(1)
    plt.plot(U, OUTPUT,label='Optimized_Interpolate');
    plt.xlabel('u')
    plt.ylabel('OUTPUT')
    plt.legend(loc='upper left')
    plt.show()
```
